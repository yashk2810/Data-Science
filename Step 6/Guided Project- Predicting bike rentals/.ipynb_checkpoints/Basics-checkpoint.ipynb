```python
import pandas as pd

bike_rentals = pd.read_csv('bike_rental_hour.csv')
bike_rentals.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.hist(bike_rentals['cnt'])
plt.show()
```

```python
bike_rentals.corr()['cnt']
```

```python
def assign_label(col):
    if col>=6 and col<12:
        return 1
    elif col>=12 and col<18:
        return 2
    elif col>=18 and col<24:
        return 3
    else:
        return 4

bike_rentals['time_label'] = bike_rentals['hr'].apply(assign_label)
print(bike_rentals['time_label'])
```

```python
train = bike_rentals.sample(frac = 0.8)
test = bike_rentals.loc[~bike_rentals.index.isin(train.index)]
```

```python
features = list(bike_rentals.columns)
features.remove('cnt')
features.remove('dteday')
features.remove('casual')
features.remove('registered')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(train[features], train['cnt'])

predictions = lr.predict(test[features])
mse = mean_squared_error(test['cnt'], predictions)
mse
```

```python
from sklearn.tree import DecisionTreeRegressor

clf = DecisionTreeRegressor(random_state=1, min_samples_leaf=5)
clf.fit(train[features], train['cnt'])

pred_dt = clf.predict(test[features])
mse_dt = mean_squared_error(test['cnt'], pred_dt)
mse_dt
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators=150, random_state=1)
rfr.fit(train[features], train['cnt'])

pred_rf = rfr.predict(test[features])
mse_rf = mean_squared_error(test['cnt'], pred_rf)
mse_rf
```
